## Programación Funcional

Pruebas con map, filter, reduce y list comprehesion

```python
def cuadrado(n):
    return n**2
```

```python
L = [2,4,5,6,3,2,16]
L
```

```python
L2 = list(map(cuadrado, L))
L2
```

```python
L2 = list(map(lambda x:x**2, L))
L2
```

```python
L2 = [cuadrado(i) for i in L]
L2
```

```python
L2 = [i**2 for i in L]
print(L2)
```

Generar diccionarios y conjuntos

```python
from random import randint
```

```python
c = {randint(0,100) for _ in range(10)}
c
```

```python
d = {chr(i):i for i in range(ord('A'),ord('Z')+1)}
d
```

```python
d = {chr(i):[randint(0,100) for _ in range(i)] for i in range(ord('A'),ord('Z')+1)}
d
```

```python
s = "hola"
L = range(10)
d = dict(zip(s,L))
d
```

```python
s = "abcdeefghijk"
L = range(40)
```

```python
d={s[i]:i for i in range(len(s))}
d
```

```python
d={s[i]:L[i] for i in range(min(len(s),len(L)))}
d
```

Filtrar las palabras de la lista que son palíndromos

```python
L = ['hola', 'ana', 'oso', 'abbccdccbba', 'abcdeba','ana', 'oso', 'abbccdccbba']
L
```

```python
L2 = list(filter(lambda c: c == c[::-1], L))
L2
```

```python
L2 = [c.upper() for c in L if c == c[::-1]]
L2
```
